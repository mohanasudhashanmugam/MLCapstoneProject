```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.preprocessing import StandardScaler
import time
from sklearn.decomposition import PCA
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
##Independent variable - x ; Dependent variable- y; Number of featues - n/k



def split_scalar(x,y):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
    x_train = x_train.fillna(x_train.median())
    x_test = x_test.fillna(x_train.median())
    sc=StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.transform(x_test)
    print(x_train.shape)
    return x_train,x_test,y_train,y_test

def pca(x_train,x_test,n):
    pca1=PCA(n_components=n)
    x_train=pca1.fit_transform(x_train)
    x_test=pca1.transform(x_test)
    explained_variance=pca1.explained_variance_ratio_
    #print(explained_variance)
    return x_train,x_test,explained_variance


def cm_prediction(classifier,x_test):
    y_pred=classifier.predict(x_test)
    
    from sklearn.metrics import confusion_matrix
    cm=confusion_matrix(y_test,y_pred)
    
    from sklearn.metrics import classification_report
    report=classification_report(y_test,y_pred)
    
    from sklearn.metrics import accuracy_score
    Accuracy=accuracy_score(y_test,y_pred)
    
    return classifier,Accuracy,report,x_test,y_test,cm

def logistic(x_train,y_train,x_test):
    from sklearn.linear_model import LogisticRegression
    classifier=LogisticRegression(random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm

def svm_linear(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='linear',random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm

def svm_NL(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='rbf',random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm

def Navie(x_train,y_train,x_test):
    from sklearn.naive_bayes import GaussianNB
    classifier=GaussianNB()
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm

def knn(x_train,y_train,x_test):
    from sklearn.neighbors import KNeighborsClassifier
    classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm

def decision(x_train,y_train,x_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm

def random(x_train,y_train,x_test):
    from sklearn.ensemble import RandomForestClassifier
    classifier=RandomForestClassifier(n_estimators=10 , criterion='entropy', random_state=0)
    classifier.fit(x_train,y_train)
    classifier,Accuracy,report,x_test,y_test,cm=cm_prediction(classifier,x_test)
    return classifier,Accuracy,report,x_test,y_test,cm

def PCA_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf):
    dataframe=pd.DataFrame(index=['n=3','n=4','n=5','n=6','n=18','n=26','n=27','n=30','n=35','n=40'],columns=['explained_variance','Logistic','SVMl','SVMnl','KNN','Naive','Decision','Random'])
    for number, index in enumerate(dataframe.index):
        #print("number, index",number, index)
        dataframe['explained_variance'][index]=pc_explained_variance[number]
        dataframe['Logistic'][index]=acclog[number]
        dataframe['SVMl'][index]=accsvml[number]
        dataframe['SVMnl'][index]=accsvmnl[number]
        dataframe['KNN'][index]=accknn[number]
        dataframe['Naive'][index]=accnav[number]
        dataframe['Decision'][index]=accdes[number]
        dataframe['Random'][index]=accrf[number]
    return dataframe


        
            
```

```python
dataset1=pd.read_csv('watson_healthcare_modified.csv',index_col=None)
df2=dataset1
```

```python
df2=pd.get_dummies(df2,drop_first=True)
df2
```

```python
x=df2.drop('Attrition_Yes',axis=1)
y=df2['Attrition_Yes']
```

```python
correlations = df2.corr()['Attrition_Yes'].abs().sort_values(ascending=False)

print("--- 🕵️‍♂️ CORRELATION CHECK FOR DATA LEAKAGE ---")
print(correlations)
```

```python
x_train,x_test,y_train,y_test=split_scalar(x,y)
```

```python
## below code is to determine no of components 
```

```python

# 4. Fit PCA with all 26 available feature dimensions
pca2 = PCA(n_components=42)
pca2.fit(x_train)

# 5. Calculate cumulative variance matrix
cumulative_variance = np.cumsum(pca2.explained_variance_ratio_)

# 6. Find the exact number of components needed for 90% variance threshold
components_for_90 = np.argmax(cumulative_variance >= 0.90) + 1

# 7. Print results to make your decision
print(f"--- PCA Variance Summary ---")
for i, crypto_variance in enumerate(cumulative_variance):
    print(f"{i+1} Components capture: {crypto_variance*100:.2f}% of total data variance")

print(f"\nRecommended: Set n_components={components_for_90} to retain 90% of your data's information.")
```

```python
pc_explained_variance=[]
acclog=[]
accsvml=[]
accsvmnl=[]
accknn=[]
accnav=[]
accdes=[]
accrf=[]
```

```python
for i in [3,4,5,6,18,26,27,30,35,40]:
    #print(x_train.shape)
    
    x_pca_train,x_pca_test,explained_variance=pca(x_train,x_test,i)
    
    # 2. Calculate the cumulative (running total) variance ratio
    cumulative_variance = np.cumsum(explained_variance)
    cumulative_pct = cumulative_variance * 100
   
    pc_explained_variance.append(cumulative_pct)
    
    classifier,Accuracy,report,x_acc_test,y_test,cm=logistic(x_pca_train,y_train,x_pca_test)
    acclog.append(Accuracy)

    classifier,Accuracy,report,x_acc_test,y_test,cm=svm_linear(x_pca_train,y_train,x_pca_test)
    accsvml.append(Accuracy)

    classifier,Accuracy,report,x_acc_test,y_test,cm=svm_NL(x_pca_train,y_train,x_pca_test)
    accsvmnl.append(Accuracy)

    classifier,Accuracy,report,x_acc_test,y_test,cm=Navie(x_pca_train,y_train,x_pca_test)
    accnav.append(Accuracy)

    classifier,Accuracy,report,x_acc_test,y_test,cm=knn(x_pca_train,y_train,x_pca_test)
    accknn.append(Accuracy)

    classifier,Accuracy,report,x_acc_test,y_test,cm=decision(x_pca_train,y_train,x_pca_test)
    accdes.append(Accuracy)

    classifier,Accuracy,report,x_acc_test,y_test,cm=random(x_pca_train,y_train,x_pca_test)
    accrf.append(Accuracy)
    
```

```python
result=PCA_Classification(acclog,accsvml,accsvmnl,accknn,accnav,accdes,accrf)
```
