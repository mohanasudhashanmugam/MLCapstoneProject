```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pickle

df=pd.read_csv("watson_healthcare_modified.csv")

df=df.drop(['EmployeeCount','StandardHours','Over18','Education'],axis=1)

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False, drop='first', dtype=int)
encoded_array = encoder.fit_transform(df[['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']])

#  Get clean names for the new columns and create a DataFrame
columns_to_encode = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime']
encoded_column_names = encoder.get_feature_names(columns_to_encode)
encoded_df = pd.DataFrame(encoded_array, columns=encoded_column_names)

# Merge back into the original dataset and drop the old raw text columns
# We use .concat() along axis=1 to join them side-by-side
final_df = pd.concat([df, encoded_df], axis=1).drop(columns=['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime'])


x=final_df.drop('Attrition',axis=1)
y=final_df['Attrition']

x=x.drop('PerformanceRating',axis=1)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)

sc=StandardScaler()
x_scaled=sc.fit_transform(x_train)



loaded_pca = PCA(n_components=26)
pca_x = loaded_pca.fit_transform(x_scaled)



```
